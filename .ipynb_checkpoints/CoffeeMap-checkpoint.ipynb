```python
import pandas as pd
import plotly.express as px

df = pd.read_csv('chocolate-ratings.csv', delimiter = ',')
```

```python
df.head()
data = df[['Broad Bean Origin', 'Rating', 'Review Date']].reset_index()
data = data.set_index(df['Review Date']).sort_index(ascending = True)
```

```python
data
```

```python
fig = px.choropleth(data, 
                    locations=data['Broad Bean Origin'],  
                    locationmode="country names", 
                    color='Rating',      
                    color_continuous_scale="bluyl",
                    range_color=(0, data['Rating'].max()), # Fixed color range scope = 'USA'
                    animation_frame="Review Date", 
                    title= "Coffee Ratings",
                    labels={'Rating':'Rating', "country":"Broad Bean Origin"}, template = 'plotly_dark')
# fig.update_layout(margin={"r":20,"t":100,"l":20,"b":20})
fig.update_layout(margin={"r":0,"t":40,"l":0,"b":0}, width=800, height=600)

fig.show()
```
