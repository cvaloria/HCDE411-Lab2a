### Exercise #1

Use the source data to construct a new data file named **“german_cars_2004.csv”** that lists cars manufactured by any **German company** (e.g., Audi, BMW, Mercedes-Benz, Porsche, and Volkswagen).

The new data should include 9 dimensions as follows:

**“Vehicle Name, Engine Size, Cyl, HP, City MPG, Hwy MPG, Weight, Wheel Base, Len, and Width).”**

The first line of the data is a header where the aforementioned dimensions are stored as a string. The actual data should be stored starting from the second row.

Save the file as **“german_cars_2004.csv”**.


```python
import csv
fin_name = "data_cars_2004.csv"  # file-in name is "fin"
fin = open (fin_name, "r")      # open "fin"
reader = csv.reader(fin)        # use the csv library to read the file

############################################################################
# Put your code here CHANGE

''' This code block loads the csv file with all cars data and selects
desired german data. The code block outputs a csv file with the filtered data '''

############################################################################
first_row = next(reader)  # read headers

german_brands = ['Audi','BMW','Mercedes-Benz','Porsche', 'Volkswagen']  # Specify german brands
selected_data = [] # initialize selected data list

for row in reader:                                     # iterate through each row 
    car_name = row[0]                                  # define row where car names exist 
    car_brand = car_name.split()[0]                    # split car brand from car names
    
    if car_brand in german_brands:                     # logic to define if current brand is german
        selected_data.append([row[0]] + row[11:])      # append desired data to row if it is a german car
        
header = [first_row[0]] + first_row[11:]               # define header as a list

output_filename = "german_cars_2004.csv"               # define file name to save extracted data to

fout = open(output_filename, "w")                      # open output_file and specify it for writing

writer = csv.writer(fout, delimiter=',', quotechar='"', 
                    quoting=csv.QUOTE_MINIMAL)         # specify parameters for writing the csv file        

writer.writerow(header)                                # write the header to the file

for row in selected_data:                              # iterate through each row of selected data to write to csv
    writer.writerow(row)

fout.close()
############################################################################
fin.close()                     # close "fin"

```

### Exercise #2

Use the source data to construct a new file named **“cars_2004_netprofit.csv”** that add attribute named “Net” to each car record.

“Net” can be calculated with simple math as follows: 
**((Retail Price–Dealer Cost)/Retail Price)*100.**

The new data should include 10 dimensions as follows:

**“Vehicle Name, Engine Size, Cyl, HP, City MPG, Hwy MPG, Weight, Wheel Base, Len, Width, and Net).”**

The first line of the data is a header where the aforementioned dimensions are stored as a comma separated string. The actual data record starts from the second row.

Save the file as **“cars_2004_netprofit.csv”**.

```python
import csv
fin_name = "data_cars_2004.csv"  # file-in name is "fin"
fin = open (fin_name, "r")      # open "fin"
reader = csv.reader(fin)        # use the csv library to read the file

############################################################################
# Put your code here

''' This code block loads the csv file with all cars data and selects desired data
The code block outputs a csv file with the filtered data and calculated 'Net' column '''


############################################################################
first_row = next(reader)                                 # read headers
data = []                                                # initialize selected data list
net = []

for row in reader:
    dealerCost = int(row[10])                                             # define column that contains dealer cost and cast as int
    retailPrice = int(row[9])                                             # define column that contains retail Price and cast as int
    net = round(100 * (retailPrice - dealerCost) / retailPrice , 2)       # calculate net
    
    data.append([row[0]] + row[11:] + [net])                              # append data matrix with desired information
    
header = [first_row[0]] + first_row[11:] + ['Net']                        # Define header

output_filename = "cars_2004_netprofit.csv"            # specify filename
fout = open(output_filename, "w")                      # open output_file and specify it for writing

writer = csv.writer(fout, delimiter=',', quotechar='"', 
                    quoting=csv.QUOTE_MINIMAL)         # specify parameters for writing the csv file        

writer.writerow(header)                                # write the header to the file

for row in data:                                       # iterate through each row of selected data to write to csv
    writer.writerow(row)
    
fout.close()
############################################################################
fin.close()                     # close "fin"
```

### Exercise #3

Use the source data to construct two data files named **“cars_2004_over50000.csv”** which stores cars that have equal or greater to 50,000 for Dealer Cost and **“cars_2004_under50000.csv”** which includes cars that have under 50,000 for Dealer Cost.

The new data should include 10 dimensions as follows (include the "Net" dimension we added in Exercise #2):

**“Vehicle Name, Engine Size, Cyl, HP, City MPG, Hwy MPG, Weight, Wheel Base, Len, Width, and Net).”**

The first line of the data is a header where the aforementioned dimensions are stored as a comma separated string. The actual data record starts from the second row.

Save the file as **“cars_2004_over50000.csv”** and **“cars_2004_under50000.csv”**.

```python
import csv
fin_name = "data_cars_2004.csv"  # file-in name is "fin"
fin = open (fin_name, "r")      # open "fin"
reader = csv.reader(fin)        # use the csv library to read the file

############################################################################
# Put your code here

""" This code section takes the cars data file and filters the data into two
separate files. The two output files contain cars data split by a Dealer Cost
of $50,000. 
"""
############################################################################

first_row = next(reader)           # read headers
value = 50000                      # define filter value
underValue = []                    # initialize list for cars under 50000 
overValue = []                     # initialize list for cars over 50000

for row in reader:
    dealerCost = int(row[10])      # specify dealer cost column and cast as integer
    retailPrice = int(row[9])      # specify retail price column and cast as integer
    net = round(100 * (retailPrice - dealerCost) / retailPrice, 2) # calculate 'net' value
    
    if dealerCost < value:                                    # logic for filtering data
        underValue.append([row[0]] + row[11:] + [net])
                          
    else:
        overValue.append([row[0]] + row[11:] + [net])
    
    
header = [first_row[0]] + first_row[11:] + ['Net']           # redefine header rows for output file

# create dictionary that includes filtered data and info for writing to csv files
data = {'over' : 
        {'filename': 'cars_2004_over50000.csv', 
         'data' : overValue, 
         'header' : header },
        
    'under' : 
        {'filename': 'cars_2004_under50000.csv', 
         'data' : underValue, 
         'header': header}}

# loop through 'data' dictionary to create csv files
for file in data:
    fout = open(data[file]['filename'], "w")
    writer = csv.writer(fout, delimiter=',', quotechar='"', quoting = csv.QUOTE_MINIMAL)
    
    for row in data[file]['data']:
        writer.writerow(row)
    fout.close()

############################################################################
fin.close() 
```

## Python JSON to CSV exercise

Source data: **[bgt_bike_and_peds.json](https://drive.google.com/open?id=1GBJunIGe5i0qbI6ZrCkzRlVPmiTgyTKt)**. 

The dataset we are working with is constructed from data.seattle.gov which present a count of traffic per hour on the Burke-Gilman trail NE 70th street, broken down by bicycles and pedestrian traffic.

### EXAMPLE
Use the source data to construct a new data named **“bgt_bike_and_peds_weekday.csv”** that shows counts of pedestrian and bicycles per each weekday. More specific file format is presented as follows:

The first line of the data is a header where the dimensions of *weekday , ped_s, ped_n, bike_s, bike_n, and total*. 

Weekday starts from 0 which means Monday and ends with 6 which means Sunday.

The actual data (in total, 7 lines) should be presented from the second row.
In checking weekday, you may find [this link](https://docs.python.org/3/library/datetime.html) to be useful.

```python
import csv
fin_name = "bgt_bike_and_peds.json"  # file-in name is "fin"
fin = open (fin_name, "r")      # open "fin"

############################################################################
# Put your code here
############################################################################
import json

counts = {}

for weekday in range(7):
    counts[weekday] = {
        "weekday": weekday,
        "ped_s": 0, 
        "ped_n": 0, 
        "bike_s": 0, 
        "bike_n": 0, 
        "total": 0
    }
    
mapping = {
      "BGT North of NE 70th Total": "total",
      "Ped South": "ped_s",
      "Ped North": "ped_n",
      "Bike North": "bike_n",
      "Bike South": "bike_s"
}

from datetime import datetime

for entry in json.load(fin):
    t = entry["Date"]
    dt_obj = datetime.strptime(t, "%m/%d/%Y %I:%M:%S %p")
    weekday = dt_obj.weekday()
    
    for term, mapped_term in mapping.items():
        val = 0
        if entry[term].strip() != '':
           val = int(entry[term])
        counts[weekday][mapped_term] += val
    #print(json.dumps(counts, indent=3))
    #input()
    
import csv
fout = open("bgt_bike_and_peds_weekday.csv", "w")
fieldnames = ["weekday", "ped_s", "ped_n", "bike_s", "bike_n", "total"]
writer = csv.DictWriter(fout, fieldnames=fieldnames)
writer.writeheader()

for weekday in range(7):
    writer.writerow(counts[weekday])

fout.close()
############################################################################
fin.close()                     # close "fin"
```

#### Exercise #4

Use the source data to construct a new data named **“bgt_bike_and_peds_month.csv”** that shows counts of pedestrian and bicycles per each month. More specific file format is presented as follows:

The first line of the data is a header where the dimensions of *month , ped_s, ped_n, bike_s, bike_n, and total*. 

Month starts from 0 which means January and ends with 11 which means December.

The actual data (in total, 12 lines) should be presented from the second row.

```python
import csv
fin_name = "bgt_bike_and_peds.json"  # file-in name is "fin"
fin = open (fin_name, "r")      # open "fin"

############################################################################
# Put your code here

"""
This code section takes source data from data.seattle.gov and sorts it by month
and outputs a csv file with the sorted data. 
"""

############################################################################
import json

counts = {}          # initialize counts dict

# initialize dictionary for storing month counts
for month in range(0,12):    
    counts[month] = {
        "month" : month,
        "ped_s": 0, 
        "ped_n": 0, 
        "bike_s": 0, 
        "bike_n": 0, 
        "total": 0}

# map variables from downloaded data to months dictionary
mapping = {
      "BGT North of NE 70th Total": "total",
      "Ped South": "ped_s",
      "Ped North": "ped_n",
      "Bike North": "bike_n",
      "Bike South": "bike_s"
}   

from datetime import datetime

# extract month from Date in data, sort into month dictionary
for entry in json.load(fin):
    t = entry["Date"]
    dt_obj = datetime.strptime(t, "%m/%d/%Y %I:%M:%S %p")
    month = dt_obj.month - 1
    
    for term, mapped_term in mapping.items():
        val = 0
        if entry[term].strip() != '':
           val = int(entry[term])
        counts[month][mapped_term] += val


fout = open("bgt_bike_and_peds_month.csv", "w")
fieldnames = ["month", "ped_s", "ped_n", "bike_s", "bike_n", "total"]
writer = csv.DictWriter(fout, fieldnames=fieldnames)
writer.writeheader()

for month in range(12):
    writer.writerow(counts[month])

fout.close()
    
############################################################################
fin.close()                     # close "fin"
```

### Exercise #5

Use the source data to construct a new data named “bgt_bike_and_peds_time.csv” that shows counts of pedestrian and bicycles per each time in a day. More specific file format is presented as follows:

The first line of the data is a header where the dimensions of time , ped_s, ped_n, bike_s, bike_n, and total.

Month starts from 0 which means 12:00 A.M. and ends with 23 which means 11:00 P.M. The actual data (in total, 24 lines) should be presented from the second row.

```python
import csv
fin_name = "bgt_bike_and_peds.json"  # file-in name is "fin"
fin = open (fin_name, "r")      # open "fin"

############################################################################
# Put your code here

"""
This code section takes source data from data.seattle.gov and sorts it by hour
and outputs a csv file with the sorted data. 
"""

############################################################################

import json

counts = {}          # initialize counts dict

# initialize dictionary for storing time counts
for hour in range(0,24):    
    counts[hour] = {
        "time" : hour,
        "ped_s": 0, 
        "ped_n": 0, 
        "bike_s": 0, 
        "bike_n": 0, 
        "total": 0}

# map variables from downloaded data to months dictionary
mapping = {
      "BGT North of NE 70th Total": "total",
      "Ped South": "ped_s",
      "Ped North": "ped_n",
      "Bike North": "bike_n",
      "Bike South": "bike_s"
}   

from datetime import datetime

# extract time from Date in data, sort into hour dictionary
for entry in json.load(fin):
    t = entry["Date"]
    dt_obj = datetime.strptime(t, "%m/%d/%Y %I:%M:%S %p")
    hour = dt_obj.hour
    
    for term, mapped_term in mapping.items():
        val = 0
        if entry[term].strip() != '':
           val = int(entry[term])
        counts[hour][mapped_term] += val

        
# specify output file name and CSV dictionary writer parameters
fout = open("bgt_bike_and_peds_time.csv", "w")
fieldnames = ["time", "ped_s", "ped_n", "bike_s", "bike_n", "total"]
writer = csv.DictWriter(fout, fieldnames=fieldnames)
writer.writeheader()

# write hour dictionary into csv
for hour in range(24):
    writer.writerow(counts[hour])

fout.close()


############################################################################
fin.close()                     # close "fin"

```
